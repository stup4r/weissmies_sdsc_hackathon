```python
import os
from dotenv import load_dotenv
from py2neo import Graph
load_dotenv()
password = os.getenv('DBPASS')

graph = Graph("bolt://localhost:7687", auth=("neo4j", password))
```

```python
query = """
MATCH (i:Issuelist {name: "Umwelt"})<-[:ABOUT]-(b:Bill)-[:CONTAINS]->(cd:`Chamber Debate`)-[:DISCUSSED_IN]->(d:Discussion)-[:DIVIDED_IN]->(p:Partition)
RETURN p
"""

partitions = graph.run(query).data()
```
